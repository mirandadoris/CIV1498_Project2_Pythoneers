# Model Comparison and Evaluation
The following notebook will compare the obtained r-squared values from the monthly and daily models.

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Summarize r-squared scores for each pollutant and each model
SO2 = {'Data':['Monthly Mean', 'Daily Mean', 'Daily Moving Average',
               'Monthly Mean', 'Daily Mean', 'Daily Moving Average'],
       'Model':['Training', 'Training', 'Training',
                'Testing', 'Testing', 'Testing'],
       'Score':[0.7227,0.2640,0.3925,
                  0.4421,0.1606,0.3022]}

TRS = {'Data':['Monthly Mean', 'Daily Mean', 'Daily Moving Average',
               'Monthly Mean', 'Daily Mean', 'Daily Moving Average'],
        'Model':['Training', 'Training', 'Training',
                'Testing', 'Testing', 'Testing'],
       'Score':[0.5962,0.3027,0.3426,
                0.2129,0.2701,0.3068]}

NO2 = {'Data':['Monthly Mean', 'Daily Mean', 'Daily Moving Average',
                'Monthly Mean', 'Daily Mean', 'Daily Moving Average'],
       'Model':['Training', 'Training', 'Training',
                'Testing', 'Testing', 'Testing'],
       'Score':[0.5027,0.3418,0.4288,
                0.6611,0.3130,0.3606]}


O3 = {'Data':['Monthly Mean', 'Daily Mean', 'Daily Moving Average',
               'Monthly Mean', 'Daily Mean', 'Daily Moving Average'],
      'Model':['Training', 'Training', 'Training',
               'Testing', 'Testing', 'Testing'],
      'Score':[0.8857,0.5679,0.6512,
               0.7381,0.5394,0.6147]}

# Create dfs
SO2_eval = pd.DataFrame(SO2)
TRS_eval = pd.DataFrame(TRS)
NO2_eval = pd.DataFrame(NO2)
O3_eval = pd.DataFrame(O3)

SO2_eval
```

```python
# create a subplot of bar graphs
fig, axs = plt.subplots(2,2, figsize=(16,10))
fig.suptitle('Training and Testing Evaluation Scores for all Models', fontsize=20)
fig.supylabel('Final Evaluation Scores', fontsize=20)

sns.barplot(ax=axs[0,0], x = "Data", y = "Score", data = SO2_eval, hue='Model')
axs[0,0].set_title('SO2 Model', fontsize=16)
sns.barplot(ax=axs[0,1], x = "Data", y = "Score", data = TRS_eval, hue='Model')
axs[0,1].set_title('TRS Model', fontsize=16)
sns.barplot(ax=axs[1,0], x = "Data", y = "Score", data = NO2_eval, hue='Model')
axs[1,0].set_title('NO2 Model', fontsize=16)
sns.barplot(ax=axs[1,1], x = "Data", y = "Score", data = O3_eval, hue='Model')
axs[1,1].set_title('O3 Model', fontsize=16)

for ax in axs.flat:
    ax.set(xlabel='', ylabel='')
    ax.set_ylim(0,1)
    ax.legend(fontsize=14)
    ax.tick_params(axis='both', labelsize=12)



plt.tight_layout()
plt.show()
```

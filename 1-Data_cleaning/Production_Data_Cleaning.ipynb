```python
import pandas as pd
```

```python
#read production data to see what are included in each column
prod_2019=pd.read_csv('prod_2019.csv')
prod_2019
```

```python
# Converting the index as date
format ='%Y%m'

prod_2019['Date'] = pd.to_datetime(prod_2019['prod_period'].astype('str'), format=format)
prod_2019
```

```python
#sum up gas production volume for each month
df_filter=prod_2019[['Date', 'gas_prod_vol_m3']]

production_monthly_2019=df_filter.groupby(by="Date").sum()
production_monthly_2019
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))

plt.plot(production_monthly_2019.index, production_monthly_2019['gas_prod_vol_m3'],'r', label = 'Monthly Sum')

plt.xlabel('Date', fontsize=14)
plt.ylabel('Gas Production Volume (m3)')

plt.title('Gas Production Volumn (m3) Trend in 2019',fontsize=16)
plt.show()
```
